```python
#Building the Matrix
for i in range(m):
    X[i,0] = 1
    X[i,1] = x[i]
    X[i,2] = x[i]**2
    Y[i] = y[i]
    
print(X)
#print(Y)
```

```python
temp1 = np.linalg.inv(np.dot(X.T,X))
temp2 = np.dot(X.T,Y)
theta = np.dot(temp1, temp2)
print(theta)
```

```python
J_temp1 = Y - np.dot(X,theta)
J = np.dot(J_temp1.T, J_temp1)/m
theta0 = theta[0,0]
theta1 = theta[1,0]
theta2 = theta[2,0]
print(f'The cost J = sum squared error = {J[0,0]:.3f}') 
```
