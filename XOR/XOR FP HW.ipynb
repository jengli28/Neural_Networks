```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    a = 1 / (1 + np.exp(-z))
    return a
```

### Begin with constructing the vectors X, W1, W2, b, based on the given wightw and biases in diagrams above ###


```python
X1 = np.array([[0,0,1,1]])
X2 = np.array([[0,1,0,1]])
Y = np.array([[0,1,1,0]])
W_l1 = np.array([[5.8,5.8],[3.7,3.7]])
W_l2 = np.array([[7.5,-8.2]])
b_l1 = np.array([[-2.5],[-5.6]])
b_l2 = np.array([[-3.4]])
```

```python
X = np.block([[X1],[X2]])
```

```python
# Forward prop layer [1]
Z_l1 = np.dot(W_l1,X) + b_l1
A_l1 = sigmoid(Z_l1)
```

```python
# Forward prop layer [2]
Z_l2 = np.dot(W_l2,A_l1) + b_l2
A_l2 = sigmoid(Z_l2)
```

```python
yhat = A_l2
yhat
```
